```
import numpy as np
import pandas as pd
import sklearn
# from sklearn.model_selection import train_test_split
dataClothes = {
    'size':['XXS','XS','S','M','M','L','L','XL','XL',np.nan,'XXL','XXXL','XXXL'],
    'color':['BLACK','WHITE',np.nan,np.nan,'BLUE','BROWN','PURPLE',np.nan,'PINK','DARKBLUE',np.nan,'GRAY','VIOLET'],
    'gender':[np.nan,np.nan,np.nan,'FEMALE',np.nan,'FEMALE','MALE','FEMALE','MALE','FEMALE','MALE','FEMALE','MALE'],
    'price':[60.0,65.0,70.0,75.0,80.0,85.0,90.0,np.nan,100.0,np.nan,np.nan,115.0,np.nan],
    'weight':[150,170,190,np.nan,230,250,270,290,310,330,350,370,390],
    'bought':['YES', 'YES','NO','NO','YES','NO','YES','YES','YES','NO','YES','NO','NO']
}
df_raw = pd.DataFrame(data=dataClothes)
df = df_raw.copy()
#1.	pomiń dane, które mają 3 lub więcej pustych wartości
df.dropna(thresh = (len(df.keys())-2), inplace=True)
print(df)
#2.	wypełnij puste miejsca w kolumnie price wartością średnią
df['price'].fillna(df['price'].mean(), inplace=True)
print(df)
 #3.	wypełnij puste miejsca w kolumnie weight za pomocą interpolacji
df['weight'].fillna(df['weight'].interpolate(), inplace=True)
print(df)
 #4.	wypełnij puste miejsca w kolumnie size, color i gender wartościami stałymi
 # (np. unisex, black, male)
df['size'].fillna('unisize', inplace=True)
df['color'].fillna('black', inplace=True)
df['gender'].fillna('male', inplace=True)
print(df)
#5.	dodaj do słownika kolumnę cost, która reprezentuje cenę jako jedną z klas: cheap, normal_price, 
#expensive; podziału ceny na te klasy dokonaj automatycznie (argument bins metody pandas.cut)
df['cost'] = pd.cut(df['price'], bins = 3, labels=('cheap', 'normal_price', 'expensive'))
df
#6.	dostosuj dane do modelu tak, by przedstawiały wartość binarną 
#(np. bought: [‘YES’, ‘NO’, ‘YES’] na bought_YES: [1,0,1])
df['bought_YES'] = df['bought'].apply(lambda x: 1 if 'YES' in x else 0)
df
#7.	podziel słownik na dane wejściowe i
# wartość wyjściową, a następnie podziel na zbiór uczący i testujący
# df1=df[['price','cost']],
# df2=df['bought_YES']
# X_train, X_test, y_train, y_test = train_test_split(df1, df2, train_size = 0.5)                                      
```
