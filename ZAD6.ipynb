```
from sklearn.datasets import load_iris
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
#Przygotowanie danych do laboratorium
raw = load_iris()
data = raw['data']
target = raw['target']
feature_names = raw['feature_names']
target_names = raw['target_names']
data = np.array(data[:,:2])
target = np.array(list(map(int,target)))
#Podział danych na uczące się i testujące 
X_train,X_test,y_train,y_test = train_test_split(data,target,test_size=0.2)
#Stworzenie obiektu klasy SVC
clf = SVC(C=0.04, kernel='linear')
clf.fit(X_train, y_train)
plot_decision_regions(X_train, y_train, clf=clf, legend=2)
y_pred = clf.predict(X_test)
plt.xlabel('Długość [cm]')
plt.ylabel('Szerokość [cm]')
plt.title('C=0.1 oraz kernel=linear')
plt.show()
#Kilka przykładów
clf1 = SVC(C=0.008, kernel='poly')
clf1.fit(X_train, y_train)
plot_decision_regions(X_train, y_train, clf=clf1, legend=2)
y_pred1 = clf1.predict(X_test)
plt.xlabel('Długość [cm]')
plt.ylabel('Szerokość [cm]')
plt.title('C=0.01 oraz kernel=poly')
plt.show()
clf2 = SVC(C=0.20, kernel='rbf')
clf2.fit(X_train, y_train)
plot_decision_regions(X_train, y_train, clf=clf2, legend=2)
y_pred2 = clf2.predict(X_test)
plt.xlabel('Długość [cm]')
plt.ylabel('Szerokość [cm]')
plt.title('C=0.01 oraz kernel=rbf')
plt.show()


clf3 = SVC(C=0.588, kernel='sigmoid')
clf3.fit(X_train, y_train)
plot_decision_regions(X_train, y_train, clf=clf3, legend=2)
y_pred2 = clf3.predict(X_test)
plt.xlabel('Długość [cm]')
plt.ylabel('Szerokość [cm]')
plt.title('C=0.244 oraz kernel=sigmoid')
plt.show()

```
